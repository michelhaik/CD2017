## Introdução

Segurança alimentar é algo de extrema importância para a vida de todos. Vamos fazer uma analogia ao carro: a gasolina/etanol é o combustível de nossos carros. Para que eles funcionem bem, é necessário esta ser de boa qualidade, garantindo um serviço de locomoção sem problemas. Com o alimento é a mesma coisa: precisamos de alimentos de boa procedência para vivermos uma vida saudável, sem problemas que impactam no nosso cotidiano.
```python
#Se usar o arquivo descompactado
dados2009=pd.read_csv('2009/DOM2009.csv',sep=',')
dados2009
#Se usar o arquivo compactado como está no dropbox púbico
#pd.read_csv('DOM2015csv.zip', compression='zip', header=0, sep=',', quotechar='"')
```

```python
#vou selecionar os estados e criar uma variavel "regiao"

dados2009.loc[(dados2009.UF >10) & (dados2009.UF<19),'Região'] = 'Norte'
dados2009.loc[(dados2009.UF >20) & (dados2009.UF<30),'Região'] = 'Nordeste'
dados2009.loc[(dados2009.UF >30) &(dados2009.UF<40),'Região'] = 'Sudeste'
dados2009.loc[(dados2009.UF >40) &(dados2009.UF<50),'Região'] = 'Sul'
dados2009.loc[(dados2009.UF >=50) &(dados2009.UF<60),'Região'] = 'Centro-Oeste'
dados2009
```

```python
#Tem geladeira

dados2009.loc[(dados2009.V0228 ==2) |(dados2009.V0228 == 4),'Geladeira'] = 'Sim'
dados2009.loc[(dados2009.V0228 ==6),'Geladeira'] = 'Não'







#Tem filtro d’água
dados2009.V0224 = dados2009.V0224.astype('category')

dados2009.V0224.cat.categories = (['Possui Filtro', 'Não possui'])

populacao_com_filtro=dados2009[dados2009.V0224 == 2]# & (dados2009.V0228 == 4) ]
populacao_com_filtro.to_csv('Pessoas_com_filtro.csv', sep=',')
#Algum morador de 18 anos ou mais de idade deixou de fazer alguma refeição porque não havia dinheiro para comprar comida


adultos_sem_refeicao=dados2009[dados2009.V2113==1]
adultos_sem_refeicao.to_csv('Adultos_sem_comida.csv', sep=',')

#Algum morador de 18 anos ou mais de idade perdeu peso porque não comeu quantidade suficiente de comida devido à falta de dinheiro para comprar comida
dados2009.V2119 = dados2009.V2119.astype('category')

dados2009.V2119.cat.categories = (['Sim', 'Não'])

adultos_magros=dados2009[dados2009.V2119=="Sim"]
adultos_magros.to_csv('Adultos_magrinhos.csv', sep=',')
```

```python
print('Pessoas com filtro de água:')
ut1 = (dados2009.V0224.value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(ut1,'\n')

print("pessoas com filtro de agua no sudeste")
sudeste=dados2009[dados2009.Região=="Sudeste"]
sud= (sudeste.V0224.value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(sud,'\n')

print("pessoas com filtro de agua no Nordeste")
nordeste=dados2009[dados2009.Região=="Nordeste"]
nord= (nordeste.V0224.value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(nord,'\n')

#####################

###
print ('moradores com mais que 18 anos que perderam peso por nao comer no nordeste')
nod= (nordeste.V2119.value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(nod,'\n')
##
print ('brasileiros com geladeira')
ut2 = (dados2009.Geladeira.value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(ut2,'\n')

###

print ('nordestinos com geladeira')
ut2 = (nordeste.Geladeira.value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(ut2,'\n')
```
