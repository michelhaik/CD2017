## Introdução

Segurança alimentar é algo de extrema importância para a vida de todos. Vamos fazer uma analogia ao carro: a gasolina/etanol é o combustível de nossos carros. Para que eles funcionem bem, é necessário esta ser de boa qualidade, garantindo um serviço de locomoção sem problemas. Com o alimento é a mesma coisa: precisamos de alimentos de boa procedência para vivermos uma vida saudável, sem problemas que impactam no nosso cotidiano. Um simples prato de feijão estragado pode nos fazer ficar algumas horas (e quem sabe dias!) indo constantimente no banheiro, atrapalhando nosso estudo, trabalho e bem estar. Nosso país, como sabemos, tem inúmeros problemas, alguns mais fáceis e outros mais complicados de resolver. Mas pense aqui comigo: a qualidade alimentar não seria um problema fácil de resolver? Dar uma lavada nas verduras, antes de cozinhar, ou colocar a água em um filtro antes de tomar, não parece coisas simples, demandando só um pouco mais de cuidado?


Das imensas filas de nosso infelizmente falho sistema de saúde, quantos são de intoxicação alimentar? Será que essa fila não poderia diminuir para cuidar de problemas mais complicados?


Minha pesquisa visa analisar os domicílios brasileiros quanto à segurança alimentar, com base nos dados da PNAD 2009 e PNAD 2013, focado no Nordeste, região mais pobre no Brasil.

```python
#Se usar o arquivo descompactado
dados2009=pd.read_csv('2009/DOM2009.csv',sep=',')
dados2009
#Se usar o arquivo compactado como está no dropbox púbico
#pd.read_csv('DOM2015csv.zip', compression='zip', header=0, sep=',', quotechar='"')
```

```python
#vou selecionar os estados e criar uma variavel "regiao"

dados2009.loc[(dados2009.UF >10) & (dados2009.UF<19),'Região'] = 'Norte'
dados2009.loc[(dados2009.UF >20) & (dados2009.UF<30),'Região'] = 'Nordeste'
dados2009.loc[(dados2009.UF >30) &(dados2009.UF<40),'Região'] = 'Sudeste'
dados2009.loc[(dados2009.UF >40) &(dados2009.UF<50),'Região'] = 'Sul'
dados2009.loc[(dados2009.UF >=50) &(dados2009.UF<60),'Região'] = 'Centro-Oeste'
dados2009
```

```python
#Criando uma variável se o sujeito possui geladeira

dados2009.loc[(dados2009.V0228 ==2) |(dados2009.V0228 == 4),'Geladeira'] = 'Sim'
dados2009.loc[(dados2009.V0228 ==6),'Geladeira'] = 'Não'







#Criando uma variável se o sujeito tem filtro d’água
dados2009.V0224 = dados2009.V0224.astype('category')

dados2009.V0224.cat.categories = (['Possui Filtro', 'Não possui'])
```

```python
print('Pessoas com filtro de água:')
ut1 = (dados2009.V0224.value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(ut1,'\n')

print("pessoas com filtro de agua no sudeste")
sudeste=dados2009[dados2009.Região=="Sudeste"]
sud= (sudeste.V0224.value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(sud,'\n')

print("pessoas com filtro de agua no Nordeste")
nordeste=dados2009[dados2009.Região=="Nordeste"]
nord= (nordeste.V0224.value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(nord,'\n')

#####################



###
#Algum morador de 18 anos ou mais de idade deixou de fazer alguma refeição porque não havia dinheiro para comprar comida



dados2009 = dados2009.rename(columns={"V2113":"refeicao"})
dados2009.refeicao = dados2009.refeicao.astype('category')

dados2009.refeicao.cat.categories =("Sim", "Não")
#dados2009.V4622.cat.categories 

print ('Algum nordestino de 18 anos ou mais de idade deixou de fazer alguma refeição porque não havia dinheiro para comprar comida')
nod= (nordeste.refeicao.value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(nod,'\n')
#################

print ('Algum brasileiro de 18 anos ou mais de idade deixou de fazer alguma refeição porque não havia dinheiro para comprar comida')
nod= (dados2009.refeicao.value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(nod,'\n')
##############
```
