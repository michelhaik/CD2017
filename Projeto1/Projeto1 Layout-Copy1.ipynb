## Introdução

Segurança alimentar é algo de extrema importância para a vida de todos. Vamos fazer uma analogia ao carro: a gasolina/etanol é o combustível de nossos carros. Para que eles funcionem bem, é necessário esta ser de boa qualidade, garantindo um serviço de locomoção sem problemas. Com o alimento é a mesma coisa: precisamos de alimentos de boa procedência para vivermos uma vida saudável, sem problemas que impactam no nosso cotidiano. Um simples prato de feijão estragado pode nos fazer ficar algumas horas (e quem sabe dias!) indo constantimente no banheiro, atrapalhando nosso estudo, trabalho e bem estar. Nosso país, como sabemos, tem inúmeros problemas, alguns mais fáceis e outros mais complicados de resolver. Mas pense aqui comigo: a qualidade alimentar não seria um problema fácil de resolver? Dar uma lavada nas verduras, antes de cozinhar, ou colocar a água em um filtro antes de tomar, não parece coisas simples, demandando só um pouco mais de cuidado?


Das imensas filas de nosso infelizmente falho sistema de saúde, quantos são de intoxicação alimentar? Será que essa fila não poderia diminuir para cuidar de problemas mais complicados?


Minha pesquisa visa analisar os domicílios brasileiros quanto à segurança alimentar, com base nos dados da PNAD 2009 e PNAD 2013, focado no Nordeste, região mais pobre no Brasil.
